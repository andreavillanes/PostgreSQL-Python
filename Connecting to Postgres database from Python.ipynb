```python
import pandas as pd
from sqlalchemy import create_engine

#Connecting to the local PostgreSQL database using Python
def connect_db():
    hostname ="localhost"
    port="5432"
    dbname="postgres"
    user = "postgres"
    pwd="class"
    # for creating connection string
    connection_str = f"postgresql://{user}:{pwd}@{hostname}:{port}/{dbname}"
    # SQLAlchemy engine
    engine = create_engine(connection_str)
    # you can test if the connection is made or not
    try:
        with engine.connect() as connection_str:
            print('Successfully connected to the PostgreSQL database')
    except Exception as ex:
        print(f'Sorry failed to connect: {ex}')
    return engine

def get_Sentiment_df(conn):
    df = pd.read_sql("""select * from "Test"."Table_Lexicon";""", con=conn)
    return df

def get_Sentiment_mosquito_df(conn):
    df = pd.read_sql("""select * from "Test"."Table_Lexicon" where word='mosquito';""", con=conn)
    return df

#main
conn = connect_db()
df_out = get_Sentiment_df(conn)
df_mos = get_Sentiment_mosquito_df(conn)
```
